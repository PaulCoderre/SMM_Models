```python
import pandas as pd
import os
```

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
```

```python
# Replace 'your_file.txt' with the actual file path
file_path = '/home/paulc600/scratch/PRMS/03_DNRC/inputs/tmax_c.day'
```

```python
# output path
output_folder_path = '/home/paulc600/scratch/PRMS/03_DNRC/inputs/tmax.day'
```

```python
# Read the text file into a DataFrame, skipping the first 3 lines
skip_lines = 3
df = pd.read_csv(file_path, skiprows=skip_lines, header=None, sep='\t')
```

```python
# Save the first 3 lines as strings
with open(file_path, 'r') as file:
    first_3_lines = [next(file) for _ in range(skip_lines)]
```

```python
df
```

```python
# Apply the conversion function to all relevant columns (skip first 6 columns)
for col in df.columns[6:]:
    df[col] = df[col].apply(celsius_to_fahrenheit)
```

```python
df
```

```python
# write output file
with open(output_folder_path, 'w') as output_file:
    for line in first_3_lines:
        output_file.write(line)
    df.to_csv(output_file, sep=' ', header=False, index=False, mode='a')
```
