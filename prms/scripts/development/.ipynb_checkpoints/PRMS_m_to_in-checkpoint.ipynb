```python
import pandas as pd
import os
```

```python
# Replace 'your_file.txt' with the actual file path
file_path = '/home/paulc600/scratch/PRMS/02_directory/inputs/precip_m.day'
```

```python
# output path
output_folder_path = '/home/paulc600/scratch/PRMS/02_directory/inputs/precip.day'
```

```python
# Read the text file into a DataFrame, skipping the first 3 lines
skip_lines = 3
df = pd.read_csv(file_path, skiprows=skip_lines, header=None, sep='\t')
```

```python
# Save the first 3 lines as strings
with open(file_path, 'r') as file:
    first_3_lines = [next(file) for _ in range(skip_lines)]
```

```python
df
```

```python
# Multiply all values in the DataFrame by conversion factor, except for the first 6 columns
for col in df.columns[6:]:
    df[col] *= 39.3701
```

```python
df
```

```python
# write output file
with open(output_folder_path, 'w') as output_file:
    for line in first_3_lines:
        output_file.write(line)
    df.to_csv(output_file, sep='\t', header=False, index=False, mode='a')
```
