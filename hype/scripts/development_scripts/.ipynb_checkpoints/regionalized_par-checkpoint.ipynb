```python
import pandas as pd
```

### Read files

```python
geodata = pd.read_csv('../../model/seperate_cal/milk_cal/data/GeoData.txt', header=0, index_col=0, sep='\t')
```

```python
# Open the text file in read mode
with open('../../geospacial/misc/stmary_seg', 'r') as file:
    # Read the lines of the file into a list
    lines = file.readlines()

# Optionally, you may strip newline characters from each line
stmary_seg = [line.strip() for line in lines]
```

```python
global_par= ['rrcs3','lp','epotdist', 'rivvel','damp'] 
```

```python
geodata
```

### Write CatchGroup (Milk= 1, St. Mary= 2)

```python
# Assuming 'original_df' is your original DataFrame
geodata_index = geodata.index
geodata_index = [int(index) for index in geodata_index]
# Create a new DataFrame with the same index
catchgroup = pd.DataFrame(index=geodata_index)

# Add the 'region' column with default values (you can adjust them as needed)
catchgroup['region'] = 1
```

```python
catchgroup
```

```python
# convert St. Mary list to int
stmary_seg = [int(item) for item in stmary_seg]
```

```python
# Iterate through the index and update the 'region' column where values match with the list
for index_value in geodata_index:
    if index_value in stmary_seg:
        catchgroup.loc[index_value, 'region'] = 2
```

```python
# Save the DataFrame to a text file without index or headers
catchgroup.to_csv('../../model/seperate_cal/stmary_cal/data/CatchGroup.txt', index=False, header=False)
```

### Write CatchDes

```python
catchdes = pd.DataFrame(index=geodata_index)
```

```python
# Add a column called 'intercept' filled with values of 1
catchdes['intercept'] = 1
```

```python
# Add columns called damp and rivvel filled with values of 0.4 and 4.95
catchdes['rrcs3'] = 0.05 # range is 0.1 to 10, value should 0.05, min 0 max 2
catchdes['lp'] = 0.75 # range is 0.5 to 1, value should 0.75, min 0.67 max 1.33
catchdes['epotdist'] = 4 # range is 2 to 6, value should be 4, min 0.5 max 1.5
catchdes['rivvel'] = 4.95 # range is 0.1 to 10, value should be 4.95, min 0.02 max 2.02
catchdes['damp'] = 0.4 # range is 0.25 to 2.25, value should be 0.4, min 0.25 max 2.25
```

```python
catchdes
```

```python
# Save the DataFrame to a text file with no index or header
catchdes.to_csv('../../model/Seperate_Cal/seperated_model/CatchDes.txt', index=False, header=False, sep='\t')
```

```python
# Manually add a header indicating the number of columns
with open('../../model/Seperate_Cal/seperated_model/CatchDes.txt', 'r+') as f:
    # Read the content of the file
    content = f.read()
    # Move the cursor to the beginning of the file
    f.seek(0, 0)
    # Write the header indicating the number of columns
    f.write(f'{len(catchdes.columns)}\n' + content)
```

### Write reg_par

```python
global_par
```

```python
# Repeat each string twice
reg_par_index = []
for string in global_par:
    reg_par_index.extend([string, string])
```

```python
# Create a DataFrame
reg_par = pd.DataFrame(index=reg_par_index, columns=['1'])
```

```python
reg_par
```

```python
# Save the DataFrame to a text file with an index and no header
reg_par.to_csv('../../model/Seperate_Cal/seperated_model/reg_par.txt', index=True, header=False)
```

```python
# Add a header to the text file indicating the number of values in string_list
with open('../../model/Seperate_Cal/seperated_model/reg_par.txt', 'r+') as f:
    content = f.read()
    f.seek(0, 0)
    f.write(f'{len(global_par)}\n' + content)
```
