```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
basin= pd.read_csv('../../model/model_versions/v_7/v_7_1/0058308.txt', index_col=0, sep='\t')
```

```python
basin = basin.iloc[1:]  # Drop the first row
```

```python
# Convert the index to a DatetimeIndex
basin.index = pd.to_datetime(basin.index)
```

```python
# Convert all columns to numeric
basin = basin.apply(pd.to_numeric, errors='coerce')
```

```python
basin['fcon'].min()
```

```python
# Plot index (dates) vs fcon
plt.figure(figsize=(14, 7))
plt.plot(basin.index, basin['fcon'], linestyle='-', color='b', alpha=1)

# Set plot labels and title
plt.xlabel('Date')
plt.ylabel('fcon')
plt.title('Date vs fcon for Subbain with icatch= 0.82')

# Rotate date labels for better readability
plt.xticks(rotation=45)


# Show the plot
plt.show()
```
