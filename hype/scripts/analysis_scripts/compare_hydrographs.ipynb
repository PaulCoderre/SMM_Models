```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
plot_1= pd.read_csv('../../model/03_HYPE/26_1_0058308.txt', index_col=0, sep='\t')
plot_2= pd.read_csv('../../model/03_HYPE/new_hds_0058308.txt', index_col=0, sep='\t')
```

```python
plot_1 = plot_1.drop(plot_1.index[0])
plot_2 = plot_2.drop(plot_2.index[0])
```

```python
# Ensure the index is datetime
plot_1.index = pd.to_datetime(plot_1.index)
plot_2.index = pd.to_datetime(plot_2.index)
```

```python
# Convert all columns to float
plot_1 = plot_1.astype(float)
plot_2 = plot_2.astype(float)
```

```python
# Replace -9999 with NaN
plot_1.replace(-9999, np.nan, inplace=True)
plot_2.replace(-9999, np.nan, inplace=True)
```

```python
# Plotting
plt.figure(figsize=(14, 7))
plt.plot(plot_1.index, plot_1['cout'], label='v26.1', alpha= 1, linewidth= 1, color= 'blue')
plt.plot(plot_2.index, plot_2['cout'], label='New HDS', alpha= 0.8, linewidth= 1, color= 'red')
#plt.plot(plot_1.index, plot_1['rout'], label='Observed', alpha= 1, linewidth= 1, color= 'green')

plt.xlabel('Date')
plt.ylabel('Streamflow (cms)')
plt.title('HYPE Comparison for MREIB')
plt.legend()
plt.grid(True)

# Add minor ticks
plt.minorticks_on()
plt.gca().xaxis.set_minor_locator(ticker.AutoMinorLocator())
plt.gca().yaxis.set_minor_locator(ticker.AutoMinorLocator())

# Save plot as PNG file
#plt.savefig('../../model/02_HYPE/plots/MREIB.png')

plt.show()
```
