```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from adjustText import adjust_text
```

### Inputs

```python
# import depression and drainage area info
depression_area= pd.read_csv('../../geospacial/misc/river_segment_analysis.csv', index_col=0)

# import HYPE results
hype_results = pd.read_csv('../../model/seperate_cal/combined_model/performance_meterics.csv',index_col=0)

# Define metrics to plot
x = 'Depression Percentage (%)'
y = 'Val KGE'
```

```python
depression_area.index = depression_area.index.astype(int)
hype_results.index = hype_results.index.astype(int)
```

```python
# Assuming df1 and df2 are the two DataFrames you want to merge
merged_results = pd.merge(hype_results, depression_area, left_index=True, right_index=True)
```

```python
merged_results
```

```python
# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(merged_results[x], merged_results[y])

# Calculate R^2
r2 = r_value**2

# Generate the line of best fit
line = slope * merged_results[x] + intercept
```

```python
# Plotting
plt.figure(figsize=(14, 7))
plt.scatter(merged_results[x], merged_results[y], color='blue', alpha=0.7, label='Data Points')
plt.plot(merged_results[x], line, color='red', label=f'$R^2$ = {r2:.2f}')

# Add labels for each point
texts = []
for i, row in merged_results.iterrows():
    texts.append(plt.text(row[x], row[y], row['Name'], fontsize=8, ha='left'))

# Adjust text to avoid overlap
adjust_text(texts, arrowprops=dict(arrowstyle="->", color='black', lw=0.5))

plt.xlabel(f'{x}')
plt.ylabel(f'{y}')
plt.title(f'{x} vs {y}')
plt.grid(True)
plt.legend()

# Save the plot as a PNG file
plt.savefig(f'../../model/seperate_cal/{y}_vs_{x}.png', dpi=300, bbox_inches='tight')

# Display the plot
plt.show()
```
