```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

### Inputs

```python
geodata= pd.read_csv('../../model/model_versions/v_6/v_6_2/GeoData.txt', sep= '\t', index_col=0)

new_params= pd.read_csv('../../model/HDS_Parameter_Validation.csv', index_col=0)

shapefile= gpd.read_file('../../geospacial/shapefiles/modified_shapefiles/Modified_SMMcat.shp')
```

### Analysis

```python
geodata.index= geodata.index.astype(int)
new_params.index= new_params.index.astype(int)
shapefile['hru_nhm']= shapefile['hru_nhm'].astype(int)
shapefile['seg_nhm']= shapefile['seg_nhm'].astype(int)
```

```python
# Create the dictionary from the 'hru_nhm' and 'seg_nhm' columns
hru_seg_dict = dict(zip(shapefile['hru_nhm'], shapefile['seg_nhm']))
```

```python
# replace new params index with river seg
new_params.index = new_params.index.map(hru_seg_dict)
```

```python
# Dropping the last column using its name
new_params = new_params.drop(columns=['Unnamed: 4'])
```

```python
# Create a DataFrame from the GeoDataFrame's desired columns
geodata_filtered = geodata[['hds_depth', 'SLC_62']]
```

```python
# Merge new_params with geo_df on their indices
merged_hds_params = new_params.merge(geodata_filtered, left_index=True, right_index=True)
```

```python
merged_hds_params
```

### Plots

```python
# Create a figure and a set of subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# First scatter plot: depressionAreaFrac vs SLC_62
axs[0].scatter(merged_hds_params['depressionAreaFrac'], merged_hds_params['SLC_62'], alpha=0.7)
axs[0].set_xlabel('depressionAreaFrac')
axs[0].set_ylabel('SLC_62')
axs[0].set_title('Scatter Plot of Depression Area Fraction vs Special Lake SLC')

# Second scatter plot: depressionDepth_m vs hds_depth
axs[1].scatter(merged_hds_params['depressionDepth_m'], merged_hds_params['hds_depth'], alpha=0.7)
axs[1].set_xlabel('depressionDepth_m')
axs[1].set_ylabel('hds_depth')
axs[1].set_title('Scatter Plot of Depression Depth vs HDS Depth')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
